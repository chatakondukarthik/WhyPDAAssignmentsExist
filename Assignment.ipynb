```python
list_of_casts_and_genres=df.loc[0:,['cast']] #subsetting required cols
VinDiesel = list_of_casts_and_genres.cast.str.contains('Vin Diesel',case=False,na=False) #checking for Vin Diesel

#This 'VinDiesel' contains full of True or False values as the above command is a condition to find Vindiesel

VinDieselCast = {}
Part_of = 0
Not_Part_of = 0
for i in VinDiesel:
    if i == True : Part_of += 1
    else :Not_Part_of += 1

#We made a library which stores the no.of movies he participated in or not
        
VinDieselCast['Part of'] = Part_of
VinDieselCast['Not Part of'] = Not_Part_of

# Creating plot 
fig = plt.figure(figsize =(10, 7)) 
plt.pie(VinDieselCast.values(),autopct = '%1.3f%%', labels = VinDieselCast.keys()) 
  
# show plot 
plt.show() 
VinDieselCast
```

So, as you can see, he participated in 22 movies from the list of movies in dataset, and we made a pie chart on it

```python
Question 6 : Perform Linear Regression on Budget vs Revenue
```

```python
x = df.loc[0:, 'budget'].values.reshape(-1,1)
y = df.loc[0:, 'revenue'].values.reshape(-1,1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  # model= 
regressor.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regressor.predict(X_train), color = 'green')
plt.title('Budget vs Revenue (Training set)')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```
